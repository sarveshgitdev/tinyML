```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated 28x28 grayscale image with pixel values from 0 to 255
image = np.random.randint(0, 256, (28, 28))
print("Original pixel range:", image.min(), "to", image.max())

# Normalize pixels to range 0-1 by dividing by 255
normalized_image = image / 255.0
print("Normalized pixel range:", normalized_image.min(), "to", normalized_image.max())

# Visualize original and normalized image side by side
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.title("Normalized Image")
plt.imshow(normalized_image, cmap='gray')
plt.colorbar()

plt.show()
```
