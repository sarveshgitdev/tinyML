```python
import matplotlib.pyplot as plt
import numpy as np

# Generic linear relationship function
def calculate_y(x, w, b):
    return w * x + b

# X values
X = [-1, 0, 1, 2, 3, 4]

# First instance parameters
w1, b1 = 10, -10
Y1 = [calculate_y(x, w1, b1) for x in X]

# Second instance parameters
w2, b2 = 2, -1
Y2 = [calculate_y(x, w2, b2) for x in X]

# Compute per-point RMSE (really just the absolute difference for each pair)
per_point_rmse = [np.sqrt((y1 - y2)**2) for y1, y2 in zip(Y1, Y2)]

# Print values
print("X\tY1\tY2\tPer-point RMSE")
for x, y1, y2, rmse in zip(X, Y1, Y2, per_point_rmse):
    print(f"{x}\t{y1}\t{y2}\t{rmse}")

# Plotting both instances
plt.scatter(X, Y1, color='blue', label='w=10, b=-10 points')
plt.plot(X, Y1, color='blue', linestyle='--', label='Y=10X-10')
plt.scatter(X, Y2, color='red', label='w=2, b=-1 points')
plt.plot(X, Y2, color='red', linestyle='-', label='Y=2X-1')

# Annotate per-point RMSE on the plot
for x, y1, y2, rmse in zip(X, Y1, Y2, per_point_rmse):
    plt.text(x, max(y1, y2) + 1, f"{rmse:.2f}", ha='center', va='bottom', fontsize=9, color='purple')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y1, Y2, and Per-point RMSE')
plt.legend()
plt.grid(True)
plt.show()
```
