```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([-1, 0, 1, 2, 3, 4])
Y_true = np.array([-3, -1, 1, 3, 5, 7])
b_fixed = -1  # You can choose your own fixed b

# Values of w to sweep
w_vals = np.linspace(-5, 12, 200)
losses = []

for w in w_vals:
    Y_pred = w * X + b_fixed
    loss = np.sqrt(np.mean((Y_pred - Y_true) ** 2))
    losses.append(loss)

plt.plot(w_vals, losses)
plt.xlabel("w")
plt.ylabel("RMSE loss")
plt.title("Parabolic Loss Curve as a Function of w")
plt.grid(True)
plt.show()
```
